```python
dados_ibge = dados_vendas[dados_vendas['Bairro'].str.upper().isin(lista_bairro_ibge)]
dados_ibge = dados_ibge.reset_index().rename(columns={'index':'index_dados_vendas'})
dados_ibge.head()
```

```python
ibge['Nome_do_subdistrito'] = ibge['Nome_do_subdistrito'].str.capitalize()
dados_ibge['Bairro'] = dados_ibge['Bairro'].str.capitalize()
quero = ibge.groupby(by=['Nome_do_subdistrito']).mean()[['V003', 'V005']].reset_index().rename(columns={'Nome_do_subdistrito':'Bairro'})
dados_ibge = dados_ibge.merge(quero, how='left', on='Bairro')
dados_ibge = dados_ibge.rename(columns={'V003':'Med_morador_bairro', 'V005':'Med_red_bairro'})
dados_ibge.head()
```

```python
dados_ibge_media = dados_ibge.groupby(by=['Bairro']).mean().reset_index()
```

```python
fig = plt.figure(figsize=(30, 8))

ax = fig.add_subplot(1, 2, 1)
ax = sns.barplot(x="Bairro", y="Med_morador_bairro", data = dados_ibge_media.sort_values(by='Med_morador_bairro', ascending=False)[:10])
ax.set(title='Média de moradores por domicílio nos bairros com casas mais populosas', xlabel='Média de moradores por casa', ylabel='Bairro')
ax.tick_params(axis='x', rotation=45)

ax = fig.add_subplot(1, 2, 2)
ax = sns.barplot(x="Bairro", y="Med_red_bairro", data = dados_ibge_media.sort_values(by='Med_morador_bairro', ascending=False)[:10])
ax.set(title='Média de renda por domicílio nos bairros com casas mais populosas', xlabel='Média da renda por casa', ylabel='Bairro')
ax.tick_params(axis='x', rotation=45)

plt.show()
```
