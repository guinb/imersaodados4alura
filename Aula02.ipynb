```python
dados_vendas.head()
```

```python
bairros = dados_vendas.groupby('Bairro')
m2_bairros = pd.DataFrame([bairros.Metragem.mean(), bairros.Valor_anuncio_float.mean()]).T
m2_bairros['Preco_m2'] = m2_bairros['Valor_anuncio_float'] / m2_bairros['Metragem']
m2_bairros = m2_bairros.reset_index()
m2_bairros.head()
```

```python
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float'] / dados_vendas['Metragem']
dados_vendas.head()
```

```python
m2_bairros2 = pd.DataFrame([dados_vendas.groupby('Bairro').Valor_m2.mean()]).T
m2_bairros2.head()
```

```python
fig = plt.figure(figsize=(30, 8))

ax = fig.add_subplot(1, 2, 1)
ax = sns.barplot(data = m2_bairros.sort_values('Preco_m2', ascending=False)[1:11], x='Bairro', y='Preco_m2')
ax.set(title='10 bairros com os maiores preços por m2', xlabel='Bairros', ylabel='Preço por m2')
for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('R$ {:.2f}'.format(y), (x.mean(), y), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=30)

ax = fig.add_subplot(1, 2, 2)
ax = sns.barplot(data = m2_bairros.sort_values('Preco_m2', ascending=True)[1:11], x='Bairro', y='Preco_m2')
ax.set(title='10 bairros com os menores preços por m2', xlabel='Bairros', ylabel='Preço por m2')
for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('R$ {:.2f}'.format(y), (x.mean(), y), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=30)

plt.show()
```


A diferença entre os bairros mais caros e baratos é bastante significativa, se pegarmos entre os 25% e 75% acho que fica mais justo.

>  Foram retirados o maior e menor bairro pois aparentam ter numeros irreais.

```python
plt.figure(figsize=(15,8))
ax = sns.relplot(x = 'Valor_m2', y='Vagas', hue='Banheiros', size='Quartos', data = dados_vendas.query('Quartos < 11 & Banheiros < 15'), sizes=(100, 1000), alpha=.3 , palette='viridis_r', height=12)
ax.set(xlim=(0,40000), ylim=(0,16),title='Relação entre valor por m2, quantidade de vagas, banheiros e quartos', xlabel='Valor por m2 em reais' , ylabel='Quantidade de vagas')
plt.show()
```

Imóveis com menos banheiros geralmente são imóveis com menos vagas também.
Quantidade de quartos e valor por m2 é melhor distribuido.

```python
plt.figure(figsize=(15,8))
ax = sns.catplot(x = 'Quartos', y='Valor_m2', data = dados_vendas.query('Banheiros < 15 & Vagas < 16'), height=10)
ax.set(ylim=(-100,50000), xlim=(-0.5,10), title='Distribuição de imóveis pela quantidade de quantos x valor por m2', xlabel='Quantidade de quartos' , ylabel='Valor por m2 em reais')
plt.show()
```

Maioria de imóveis com 3 e 4 quartos, com valor do m2 até 20 mil reais. 

```python
plt.figure(figsize=(15,8))
ax = sns.displot(x = 'Valor_m2', hue='Quartos', data = dados_vendas.query('Quartos < 11 & Banheiros < 20 & Vagas < 20'), kind='kde', multiple='fill', palette='hsv', height=10)
ax.set(xlim=(0,50000),title='Densidade de imóveis por valor do m2 e quartos', xlabel='Valor por m2 em reais', ylabel='Densidade')
plt.show()
```

*  Domínio dos imóveis de 3 e 4 quartos nas faixas de valor por m2 que mais tem imóveis.
*  Imóveis acima de 6 quartos são pouquissimos.
*  A partir de 35 mil reais por m2 os dados ficam bagunçados.
